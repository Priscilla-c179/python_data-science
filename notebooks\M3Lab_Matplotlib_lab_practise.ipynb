## Set up the environment

### Task: import packages and rename them accordingly

````python
import pandas as pd
import numpy as np
import matplotlib .pyplot as plt
import seaborn as sns
````

## Load the dataset

````python
data=sns.load_dataset('anscombe')
data.head()
````

## Play with distribution plots
You can show the data distribution for some attributes interest you.

````python
plt.scatter(data['x'],data['y'])
````

## Play with relational plots
You can show how two or more attributes are associated.

````python
plt.plot(data['x'],label='X')
plt.plot(data['y'],label='Y')
plt.legend()
plt.grid()
plt.show()
````

## Play with other plots
You can play with other plots we learned here. Be as creative as you can and see how the visualization helps you to understand the data better.

````python
#set colors based on x values

##colors= ['red' if label =='x' else 'blue' for lable in data['x']]
#no of data points
n=len(data)
# get X position for each bar group
indices= np.arange(n)
# get width of each bar
bar_width=0.6
# plot x and y bars
# put x bar slightly to the left by indices- bar_width
plt.bar(indices - bar_width/2, data['x'], width=bar_width, color='red', label='x')
#put y bar to the right by indices + barwidth to avoid x,y overlapping
plt.bar(indices + bar_width/2, data['y'], width=bar_width, color='blue', label='y')
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.title('X and Y values')
plt.legend()
plt.show()
````
