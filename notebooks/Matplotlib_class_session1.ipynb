```python
import matplotlib.pyplot as plt
import pandas as pd  # for tabular data
import numpy as np # multi dimension array
```

```python
x=np.arange(10)
y=np.arange(0,1000,100)

# create a plot
plt.plot(x,y)

```

x is under horizontal axis and y is under vertical axis. x,y have perfect
relationship.when x increases y increases.

Numpy can generate random numbers which can follow certain distribution as we specify

```python
# float with normal distribution with mean of 0.5,std 0.1,and a shape of 1 dimension

x=np.random.normal(0.5,0.1,(100))
#sort list from least to greatest
x.sort()
y=x**2
#print(x)
#print(y)
plt.plot(x,y)

```

the relationship is a curve which follows y=x**2

```python
# use a specific mark on datapoints like 'o'
plt.plot(x,y,'o')
```

```python
x=np.arange(100)
y=x**3
plt.plot(x,y,'1')
```

```python
y=np.array([1,3,13])
line=['-','--','-.',':']
#color=['r','g','b','y','k','w','c','m']
for l in line:
  print(l)
  # l='o'+l+'r' # This line was overwriting the variable l
  plt.plot(y,'o'+l+'r') # Moved the string concatenation here
  plt.show()
```

```python
color=['r','g','b','y','k','w','c','m']
for c in color:
  print(c)
  c='o:' + c
  plt.plot(y,c)
  plt.show()
```

```python
# ms means marker size
plt.plot(y,marker='^',ms=30)
```

```python
# check multiple relationship
x=np.arange(10)
y1=x*2 # multiplication
y2=x**2  # Exponentiation (power).
plt.plot(y1)
plt.plot(y2)
```

```python
#make plot more professional add labels
plt.plot(x, y1,label='double')
plt.plot(x, y2,label='square')
plt.xlabel("Xs")
plt.ylabel("Ys")
plt.grid()
plt.legend(['1st','2nd'])     # help use to understand graph easier.
```

```python
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
sizes=np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])
plt.scatter(x,y, c=colors,cmap='hot',s=sizes,alpha=0.5)
plt.colorbar()
```

```python
x=[0.66,0.25,0.09]
plt.pie(x,labels=['2/3','1/5','others'])
plt.legend()
```

```python
# give emphasis on what matters by explode
x=[0.66,0.25,0.09]
plt.pie(x,labels=['2/3','1/5','others'],explode=[0.0,0.2,0])
plt.legend()
```
